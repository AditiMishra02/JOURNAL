```python
# ILPD

import numpy as np
import csv
from statistics import mode

l1 = []
l2 = []
l3 = []

with open('Indian Liver Patient Dataset (ILPD).csv', newline = '') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)
    #print(header)
    for line in reader:
        tot_bilirubin = line[2]
        tot_protein = line[4]
        albumin = line[5]
        if tot_bilirubin == 'nan':
            pass
        else:
            tot_bilirubin = float(tot_bilirubin)
            l1.append(tot_bilirubin)

        if tot_protein == 'nan':
            pass
        else:
            tot_protein = int(tot_protein)
            l2.append(tot_protein)
        
        if albumin == 'nan':
            pass
        else:
            albumin = int(albumin)
            l3.append(albumin)

print('MEAN OF THE LIST 1 = ', np.mean(l1))
print('MEDIAN = ', np.median(l1))
print('MODE = ', mode(l1))
print('MAXIMUM = ', max(l1))
print('MINIMUM = ', min(l1))
print()
print('MEAN OF THE LIST 2 = ', np.mean(l2))
print('MEDIAN = ', np.median(l2))
print('MODE = ', mode(l2))
print('MAXIMUM = ', max(l2))
print('MINIMUM = ', min(l2))
print()
print('MEAN OF THE LIST 3 = ', np.mean(l3))
print('MEDIAN = ', np.median(l3))
print('MODE = ', mode(l3))
print('MAXIMUM = ', max(l3))
print('MINIMUM = ', min(l3))

with open('Indian Liver Patient Dataset (ILPD).csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    list1 = []
    for row in reader:
        list1.append(row)
with open('Indian Liver Patient Dataset (ILPD).csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    row_count = len(list(reader))

i = 0
train = []
test = []
print(row_count)
x = int(row_count * 0.8)
    # print('TEST CASE = ', x)
for num in range(row_count):
    if i < x:
        train.append(list1[num])
    else:
        test.append(list1[num])
    i = i + 1
print('NO. OF TRAINING DATASET = ', len(train))
print('NO. OF TESTING DATASET', len(test))

```
